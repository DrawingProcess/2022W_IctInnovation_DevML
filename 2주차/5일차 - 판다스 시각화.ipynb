## 연습문제
for 문을 사용해 iris 데이터의 12개의 그림을 모두 그려보자.

```python
for f0 in iris['feature_names']: # irisdt에서는 columns 이름을 뽑아야 함
    for f1 in iris['feature_names']:
        if f0 == f1:
            continue
        plt.scatter(irisdt[f0], irisdt[f1], c=iris.target)
        plt.title(f"{f0} vs {f1}")
        plt.show()
```

# 타이타닉 생존자 분석 연습 문제
피봇 테이블을 사용해 데이터를 분석해봅시다.
1. 가장 많이 살아 남은 성별은 남녀 중 어느 쪽인가요?
2. 가장 많이 살아 남은 Class는 어디 인가요? (예: 1,2,3)
3. 가장 많이 살아 남은 연령대는 어느 연령대 인가요? (예: 30대)
4. 두 번째로 가장 많이 살아 남은 연령대는 어느 연령 대 인가요? (예: 30대)



```python
titanic_df = pd.read_csv('titanic_train.csv')
titanic_df
```

```python
# 1
titanic_df['Survived'].groupby(titanic_df['Sex']).mean()
```

```python
titanic_df.pivot_table(index=['Sex'])['Survived']
```

```python
# 2
titanic_df.pivot_table(index=['Pclass'], aggfunc=np.sum)['Survived']
```

```python
titanic_df['Survived'].groupby(titanic_df['Pclass']).sum()
```

```python
# 3, 4
ages = []
for index, row in titanic_df.iterrows():
    ages.append((row['Age']//10) * 10)

titanic_df['ages'] = ages

titanic_df['Survived'].groupby(titanic_df['ages']).sum().sort_values(ascending=False)
```

```python
titanic_df['Survived'].groupby(titanic_df['ages']).sum().plot(kind='bar')
```

```python
def age_to_ages(df):
    return (df['Age']//10) * 10

titanic_df.apply(age_to_ages, axis=1)
```

# 머신러닝 맛보기

```python
from sklearn import datasets

iris = datasets.load_iris()
irisdt = pd.DataFrame(iris['data'], columns=iris['feature_names'])
irisdt
```

```python
from sklearn import neighbors

x = iris['data']
y = iris['target']

knn = neighbors.KNeighborsClassifier() # knn 알고리즘은 분류를 하는 기능을 가진다.
knn.fit(x, y) # 특징과 결과를 가지고 훈련 시킨다. 
```

```python
# 임의의 x를 넣어서 답을 요구하면 모델은 그 답을 준다.
knn.predict([[5.1, 3.5, 1.4, 0.2], [0,0,0,0]])
```

```python
knn.score(x, y) # 모델로 prediction을 수행하고 y 채점한 결과
```

# Seaborn을 활용한 데이터 시각화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
```

```python
tips_df = sns.load_dataset("tips")
tips_df
```

## scatter 그리기

```python
sns.relplot(data=tips_df, x="total_bill", y="tip")
```

```python
sns.relplot(data=tips_df, x="total_bill", y="tip", hue='smoker') # 색상으로 세 번째 특징인 흡연 여부
```

```python
sns.relplot(data=tips_df, x="total_bill", y="tip", hue='smoker', style='time') # 모양으로 네 번째 특징인 식사 인원
```

```python
sns.relplot(data=tips_df, x="total_bill", y="tip", hue="size") # 색상으로 숫자를 표현
```

```python
sns.relplot(data=tips_df, x="total_bill", y="tip", size="size") # 크기로 숫자를 표현
```

```python
_dict = dict(time=np.arange(500), value=np.random.randn(500).cumsum())
timeseries_df = pd.DataFrame(_dict)
timeseries_df
```

```python
# 시간에 따른 값이 어떻게 변화되는 라인 차트로 그림
sns.relplot(data=timeseries_df,x='time', y='value', kind='line')
```
